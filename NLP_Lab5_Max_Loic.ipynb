```python
def preprocess(resp):
  resp=resp.lower()
  resp="eos " + resp
  resp=resp.replace("."," eos")
  return resp

def generate_tokens(resp):
  tokens = word_tokenize(resp)
  return tokens

def generate_tokens_freq(tokens):
  freq_dictionary = {}
  for i in tokens:
    freq_dictionary[i]=0

  for i in tokens:
    freq_dictionary[i]+=1
  return freq_dictionary

def generate_ngrams(tokens,k):
  ngrams=[]
  i=0
  while(i<len(tokens)):
    ngrams.append(tokens[i:i+k])
    i=i+1
  ngrams=ngrams[:-1]
  return ngrams

def generate_ngram_freq(ngram):
  ngram_dict = {}
  for i in ngram:
    st=" ".join(i)
    ngram_dict[st]=0
  for i in ngram:
    st=" ".join(i)
    ngram_dict[st]+=1
  return ngram_dict

def find_ngrams(s,ngrams):
  try:
    return ngrams[s]
  except:
    return 0

def print_probability_table(distinct_tokens, freq_dictionary, ngrams):
   n=len(distinct_tokens)
   ngrams=[[]*n for i in range(n)]
   for i in range(n):
     denominator = freq_dictionary[distinct_tokens[i]]
     for j in range(n):
       numerator = find_ngrams(distinct_tokens[i]+" "+distinct_tokens[j],ngrams)
       ngrams[i].append(float("{:.3f}".format(numerator/denominator)))
   return ngrams
```
